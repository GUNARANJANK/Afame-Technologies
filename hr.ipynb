```python
import pandas as pd

# Load the data
hr_data_path = 'D:\\Afame\\HR Data.csv'  # Update with your file path
hr_df = pd.read_csv(hr_data_path)

# Display the first few rows of the HR data
print("Original Data:")
print(hr_df.head())

# 1. Remove unnecessary columns (example: 'EmployeeNumber' and 'StandardHours')
columns_to_remove = ['EmployeeNumber', 'StandardHours']
hr_df = hr_df.drop(columns=columns_to_remove)

# 2. Rename columns (example: 'Attrition' to 'Turnover')
columns_rename = {
    'Attrition': 'Turnover',
    'JobRole': 'Position',
    'MonthlyIncome': 'Monthly_Salary'
}
hr_df = hr_df.rename(columns=columns_rename)

# 3. Eliminate redundant entries (remove duplicates)
hr_df = hr_df.drop_duplicates()

# 4. Sanitize specific columns (example: strip whitespace from strings)
string_columns = hr_df.select_dtypes(include=['object']).columns
for col in string_columns:
    hr_df[col] = hr_df[col].str.strip()

# 5. Eliminate NaN values
hr_df = hr_df.dropna()

# 6. Check for additional necessary changes
# Example: Converting 'Age' to categorical data
hr_df['Age_Group'] = pd.cut(hr_df['Age'], bins=[18, 25, 35, 45, 55, 65], labels=['18-25', '26-35', '36-45', '46-55', '56-65'])

# Display the cleaned data
print("Cleaned Data:")
print(hr_df.head())

# Save the cleaned data to a new CSV file in the current directory
cleaned_data_path = 'Cleaned_HR_Data.csv'
hr_df.to_csv(cleaned_data_path, index=False)
print(f"Cleaned data saved to {cleaned_data_path}")
```
